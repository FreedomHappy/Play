```python
import requests
import json
url = ('https://newsapi.org/v2/top-headlines?'
       'country=us&'
       'apiKey=574412636f2340ddb50bf51fe85cebf2')
response = requests.get(url)
result = json.dumps(response.json(),sort_keys=True,indent=4,separators=(", ",": "))
print(type(response))
print(result)
```
